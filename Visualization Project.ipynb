----

**Из-за особенностей отображения рейтингов в виде звёзд на страницах HTML, истинные рейтинги пользователей могут немного отличаться от рейтингов, которые показываются пользователю.**

**График KDE plot, который отображает распределение отображаемых рейтингов (STARS) и истинных рейтингов на основе голосов пользователей (RATING).**

```python
plt.figure(figsize=(10,4),dpi=100)
sns.kdeplot(data=has_vote,x='RATING',clip=[0,5],fill=True,label='True Rating')
sns.kdeplot(data=has_vote,x='STARS',clip=[0,5],fill=True,label='Stars displayed')
plt.legend(loc=(1.05,0.5))
```

**Подсчет разницы в численном виде.**

```python
has_vote['STARS_DIFF'] = (has_vote['STARS'] - has_vote['RATING']).round(2)
has_vote
```

**График count plot для отображения того, сколько раз встречается то или иное значение разницы между STAR и RATING:**

```python
plt.figure(figsize=(12,4),dpi=150)
sns.countplot(x='STARS_DIFF',data=has_vote,palette='magma')
```

**На этом графике видно, что один из фильмов имеет разницу в 1 звезду между отображаемым рейтингом и истинным рейтингом.**

```python
has_vote[has_vote['STARS_DIFF'] == 1.0]
```

### Рейтинги от Rotten Tomatoes

Здесь два набора рейтингов - рейтинги, опубликованные официальными критиками, и пользовательские рейтинги. 

**График scatterplot, изображающий для Rotten Tomatoes связь между рейтингами от критиков и пользовательскими рейтингами.**

```python
plt.figure(figsize=(12,4))
sns.scatterplot(data=all_sites,x='RottenTomatoes', y='RottenTomatoes_User')
plt.xlim(0,100)
plt.ylim(0,100)
```

Полученные значения Rotten_Diff - это рейтинг критиков минус рейтинг пользователей. Близкие к нулю значения говорят о том, что критики и пользователи согласны друг с другом. Большие положительные значения - когда критики оценили фильм намного выше, чем пользователи. А большие отрицательные значения - наоборот, пользователи оценили фильм намного выше, чем критики.


```python
all_sites['Rotten_Diff']=all_sites['RottenTomatoes']-all_sites['RottenTomatoes_User']
all_sites
```

**Средняя абсолютная разницу между рейтингами от критиков и пользовательскими рейтингами для Rotten Tomatoes.**

```python
all_sites['Rotten_Diff'].apply(abs).mean()
```

**График распределения разницы между рейтингами от критиков и пользовательскими рейтингами для Rotten Tomatoes.**

```python
plt.figure(figsize=(10,4),dpi=100)
sns.histplot(x='Rotten_Diff', data=all_sites, kde=True,bins=25)
```

**График распределения *абсолютных значений* разницы между рейтингами от критиков и пользовательскими рейтингами для Rotten Tomatoes.**

```python
plt.figure(figsize=(10,4),dpi=200)
sns.histplot(x=all_sites['Rotten_Diff'].apply(abs),bins=25,kde=True)
```

**5 фильмов которые в среднем были выше всего оценены пользователями, по сравнению с оценками от критиков:**

```python
all_sites[['FILM','Rotten_Diff']].nsmallest(5,'Rotten_Diff')
```

**5 фильмов, которые в среднем были выше всего оценены критиками, по сравнению с оценками от пользователей.**

```python
all_sites[['FILM','Rotten_Diff']].nlargest(5,'Rotten_Diff')
```

## Сравнение рейтингов Fandango с рейтингами других компаний


```python
df = pd.merge(fandango,all_sites,on='FILM',how='inner')
```

```python
df.info()
```

```python
df.head()
```

### Нормализация колонок до диапазона 0-5 в соответствии с колонками STARS и RATINGS от Fandango


```python
df['RT_Norm'] = np.round(df['RottenTomatoes']/20,1)
df['RTU_Norm'] =  np.round(df['RottenTomatoes_User']/20,1)
df['Meta_Norm'] =  np.round(df['Metacritic']/20,1)
df['Meta_U_Norm'] =  np.round(df['Metacritic_User']/2,1)
df['IMDB_Norm'] = np.round(df['IMDB']/2,1)
```

```python
df.head()
```

**Создание DataFrame с названием norm_scores, в котором только нормализованные рейтинги.**

```python
norm_scores=df[['STARS','RATING','RT_Norm','RTU_Norm','Meta_Norm','Meta_U_Norm','IMDB_Norm']]
```

```python
norm_scores.head()
```

### Сравнение распределений рейтингов от разных компаний


```python
def move_legend(ax, new_loc, **kws):
    old_legend = ax.legend_
    handles = old_legend.legendHandles
    labels = [t.get_text() for t in old_legend.get_texts()]
    title = old_legend.get_title().get_text()
    ax.legend(handles, labels, loc=new_loc, title=title, **kws)
```

```python
fig, ax = plt.subplots(figsize=(15,6),dpi=150)
sns.kdeplot(data=norm_scores,clip=[0,5],shade=True,palette='Set1',ax=ax)
move_legend(ax, "upper left")
```

**График KDE plot, сравнивающий распределение рейтингов Rotten Tomatoes от критиков и распределение рейтингов STARS от Fandango.**

```python
fig, ax = plt.subplots(figsize=(15,6),dpi=150)
sns.kdeplot(data=norm_scores[['RT_Norm','STARS']],clip=[0,5],shade=True,palette='Set1',ax=ax)
move_legend(ax, "upper left")
```

**График histplot, сравнивающий все нормализованные рейтинги.**

```python
plt.subplots(figsize=(15,6),dpi=150)
sns.histplot(norm_scores,bins=50)
```


### Как различные компании оценивают фильмы с наименьшими рейтингами

**График clustermap для всех нормализованных рейтингов.**

```python
sns.clustermap(norm_scores,col_cluster=False)
```

**10 фильмов с наихудшими рейтингами по версии Rotten Tomatoes Critic.**

```python
norm_films=df[['STARS','RATING','RT_Norm','RTU_Norm','Meta_Norm','Meta_U_Norm','IMDB_Norm','FILM']]
```

```python
norm_films.nsmallest(10,'RT_Norm')
```

**Визуализация распределения рейтингов от всех компаний для 10 фильмов с наихудшими рейтингами.**

```python
plt.figure(figsize=(15,6),dpi=150)
worst_films = norm_films.nsmallest(10,'RT_Norm').drop('FILM',axis=1)
sns.kdeplot(data=worst_films,clip=[0,5],shade=True,palette='Set1')
plt.title("Рейтинги 10 худших фильмов, рассмотренных критиками RT");
```
